### 1. Introduction

* 생성적 적대 신경망은 계산 불가능한 확률적 계산을 회피함으로서 생성 모델의 프레임 워크를 대체
   * 마르코프 체인 연산이 불필요, 역전파 연산만 사용, 별다른 추측 필요 x
   
   
* unconditioned 생성 모델에서는 데이터가 생성되는 종류(mode)를 조정할 방법이 없음

=> 추가적인 정보에 대한 conditioning을 통해 데이터 생성 과정을 제어할 수 있음
   * conditioning(조건 설정)은 class label에 기반

### 2. Related Works

### 2.1 Multi-modal Learning for Image Labelling

* 지도학습에서의 놀라운 성과에도 불구하고, 굉장히 많은 수의 출력 category를 수용할 수 있도록 모델을 scaling하는 것은 어려운 과제 \
  => 다른 modality들의 추가적인 정보를 이용하는 것으로 일부 해결 가능
  
  
* 많은 태스크가 입력-출력의 1:1 mapping에 주목 해왔지만, 현실의 문제에서는 확률적인 일대다 mapping case가 많이 존재함 \
  => 조건부 확률적 생성모델을 사용 (입력 : conditioning variable / 출력 : conditional predictive distribution)

## 3. Conditional Adversarial Nets

### 3.1 Generative Adversarial Nets

* GAN은 크게 데이터 분포를 입력받아 실제에 가깝게 데이터를 생성하는 G, 입력받은 데이터가 진짜 데이터인지 G가 생성한 것인지 판별하는 D로 구성
   * 둘 다 비선형성을 띄는 mapping함수일 수 있음  ex)MLP
   
   
* G와 D는 동시에 훈련됨
   * G : min log(1-D(G(Z)))
   * D : min log(D(X))
   
 

### 3.2 Conditional Adversarial Nets

* G와 D가 추가적인 정보 y에 대해 conditioned될 수 있으면 조건부 생성모델을 만들 수 있음
   * y는 class label이나 다른 modality의 데이터등 auxiliary information
   * y를 G, D의 input layer에 추가로 입력함으로서 가능
   
   
* G : prior input noise p(z)와 y가 합쳐진 형태가 됨  /  D : x, y가 입력으로
